```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getNthframe(video, frame_number):
    cap = cv.VideoCapture(video)
    total_frames = int(cap.get(cv.CAP_PROP_FRAME_COUNT))

    if frame_number > total_frames or frame_number < 0:
        print("Error: Requested frame out of bounds")

    cap.set(cv.CAP_PROP_POS_FRAMES, frame_number)

    ret,frame = cap.read()
    return frame
```

```python
img = getNthframe("shaky cam footage.mov", 23)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(gray, cmap="gray")
plt.axis("off")

corners = cv.goodFeaturesToTrack(gray,300,0.01,10)
corners = corners.astype(int)
 
for i in corners:
    x,y = i.ravel()
    cv.circle(img,(x,y),15,255,-1)

plt.figure() 
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
#optical flow
lk_params = dict(
    winSize  = (21, 21),      # search window around point
    maxLevel = 3,
    criteria = (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 30, 0.01)
)

def track_frame_pair(frame1, frame2, corners):
    gray1 = cv.cvtColor(frame1, cv.COLOR_BGR2GRAY)
    gray2 = cv.cvtColor(frame2, cv.COLOR_BGR2GRAY)

    p0 = corners.astype(np.float32).reshape(-1, 1, 2)
    p1, status, err = cv.calcOpticalFlowPyrLK(gray1, gray2, p0, None, **lk_params)

    # only keep points frames agree on
    mask = status.ravel() == 1
    p0_good = p0[mask]
    p1_good = p1[mask]

    return p0_good, p1_good, mask

def get_consecutive_frames(video, frame_i):
    cap = cv.VideoCapture(video)
    cap.set(cv.CAP_PROP_POS_FRAMES, frame_i)
    ret1, f1 = cap.read()
    ret2, f2 = cap.read()
    cap.release()
    if not (ret1 and ret2):
        raise ValueError(f"Couldn't read frames {frame_idx}, {frame_idx+1}")
    return f1, f2
```

```python
VIDEO = "shaky cam footage.mov"

frame_a, frame_b = get_consecutive_frames(VIDEO, 200)

gray_a = cv.cvtColor(frame_a, cv.COLOR_BGR2GRAY)
corners = cv.goodFeaturesToTrack(gray_a, 300, 0.01, 10)

p0, p1, mask = track_frame_pair(frame_a, frame_b, corners)
vis = frame_a.copy()
for (x0, y0), (x1, y1) in zip(p0.reshape(-1,2), p1.reshape(-1,2)):
    x0,y0,x1,y1 = int(x0),int(y0),int(x1),int(y1)
    cv.arrowedLine(vis, (x0,y0), (x1,y1), (0,255,0), 2, tipLength=0.4)
    cv.circle(vis, (x0,y0), 4, (0,200,255), -1)

import matplotlib.pyplot as plt
vis_rgb = cv.cvtColor(vis, cv.COLOR_BGR2RGB)
plt.figure(figsize=(10,6))
plt.imshow(vis_rgb)
plt.axis("off")
plt.show()
```

```python
def get_transform(p0, p1):
    matrix, inliers = cv.estimateAffinePartial2D(p0, p1, method=cv.RANSAC,
                                                  ransacReprojThreshold=3)
    if matrix is None:
        print("estimation failed, using identity")
        return np.eye(2, 3, dtype=np.float32), 0

    n_inliers = int(inliers.sum()) if inliers is not None else 0
    return matrix, n_inliers


def decompose_transform(matrix):
    dx    = matrix[0, 2]
    dy    = matrix[1, 2]
    angle = np.arctan2(matrix[1, 0], matrix[0, 0]) * (180 / np.pi)
    scale = np.sqrt(matrix[0, 0]**2 + matrix[1, 0]**2)
    return dx, dy, angle, scale

matrix, n_inliers = get_transform(p0_good, p1_good)
dx, dy, angle, scale = decompose_transform(matrix)

print(f"Inliers      : {n_inliers} / {len(p0_good)}")
print(f"Translation  : dx={dx:.3f}px  dy={dy:.3f}px")
print(f"Rotation     : {angle:.4f} degrees")
print(f"Scale        : {scale:.5f}  (1.0 = no zoom)")
```

```python
def get_all_transforms(video_path):
    cap = cv.VideoCapture(video_path)
    total = int(cap.get(cv.CAP_PROP_FRAME_COUNT))
    
    lk_params = dict(winSize=(21,21), maxLevel=3,
                     criteria=(cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 30, 0.01))

    transforms = []
    ret, prev_frame = cap.read()
    prev_gray = cv.cvtColor(prev_frame, cv.COLOR_BGR2GRAY)

    for i in range(1, total):
        ret, curr_frame = cap.read()
        if not ret:
            break

        curr_gray = cv.cvtColor(curr_frame, cv.COLOR_BGR2GRAY)

        # redetect corners every frame
        corners = cv.goodFeaturesToTrack(prev_gray, 300, 0.01, 10)

        if corners is None:
            transforms.append((np.eye(2, 3, dtype=np.float32), 0, 0, 0, 1))
            prev_gray = curr_gray
            continue

        p0 = corners.reshape(-1, 1, 2).astype(np.float32)
        p1, status, _ = cv.calcOpticalFlowPyrLK(prev_gray, curr_gray, p0, None, **lk_params)

        mask = status.ravel() == 1
        matrix, _ = get_transform(p0[mask], p1[mask])
        dx, dy, angle, scale = decompose_transform(matrix)

        transforms.append((matrix, dx, dy, angle, scale))
        prev_gray = curr_gray

        if i % 100 == 0:
            print(f"  {i}/{total} frames processed")

    cap.release()
    return transforms


transforms = get_all_transforms(VIDEO)
print(f"\nTotal transforms: {len(transforms)}")

# summary of shake
dxs    = [t[1] for t in transforms]
dys    = [t[2] for t in transforms]
angles = [t[3] for t in transforms]

print(f"dx    range: {min(dxs):.2f} to {max(dxs):.2f}px  (stdev {np.std(dxs):.3f})")
print(f"dy    range: {min(dys):.2f} to {max(dys):.2f}px  (stdev {np.std(dys):.3f})")
print(f"angle range: {min(angles):.4f} to {max(angles):.4f} deg  (stdev {np.std(angles):.5f})")
```

```python
trajectory_x = np.cumsum([t[1] for t in transforms])
trajectory_y = np.cumsum([t[2] for t in transforms])
trajectory_a = np.cumsum([t[3] for t in transforms])

frames = np.arange(len(trajectory_x))

fig, axes = plt.subplots(3, 1, figsize=(12, 8), sharex=True)
fig.suptitle("Raw camera trajectory (before smoothing)", fontsize=14)

axes[0].plot(frames, trajectory_x, color='steelblue', linewidth=1)
axes[0].set_ylabel("X position (px)")
axes[0].axhline(0, color='gray', linewidth=0.5, linestyle='--')

axes[1].plot(frames, trajectory_y, color='coral', linewidth=1)
axes[1].set_ylabel("Y position (px)")
axes[1].axhline(0, color='gray', linewidth=0.5, linestyle='--')

axes[2].plot(frames, trajectory_a, color='mediumseagreen', linewidth=1)
axes[2].set_ylabel("Angle (degrees)")
axes[2].axhline(0, color='gray', linewidth=0.5, linestyle='--')
axes[2].set_xlabel("Frame number")

plt.tight_layout()
plt.show()
```

```python
def moving_average(curve, radius):
    window_size = 2 * radius + 1
    kernel = np.ones(window_size) / window_size
    curve_padded = np.pad(curve, (radius, radius), 'edge')
    curve_smoothed = np.convolve(curve_padded, kernel, mode='same')
    curve_smoothed = curve_smoothed[radius:-radius]
    return curve_smoothed


def smooth_trajectory(trajectory_x, trajectory_y, trajectory_a, radius=15):
    return (
        moving_average(trajectory_x, radius),
        moving_average(trajectory_y, radius),
        moving_average(trajectory_a, radius)
    )


def get_corrections(trajectory_x, trajectory_y, trajectory_a,
                    smoothed_x, smoothed_y, smoothed_a):

    return (
        smoothed_x - trajectory_x,
        smoothed_y - trajectory_y,
        smoothed_a - trajectory_a
    )
```

```python
trajectory_x = np.cumsum([t[1] for t in transforms])
trajectory_y = np.cumsum([t[2] for t in transforms])
trajectory_a = np.cumsum([t[3] for t in transforms])

SMOOTH_RADIUS = 50

smoothed_x, smoothed_y, smoothed_a = smooth_trajectory(
    trajectory_x, trajectory_y, trajectory_a, radius=SMOOTH_RADIUS
)

correction_x, correction_y, correction_a = get_corrections(
    trajectory_x, trajectory_y, trajectory_a,
    smoothed_x, smoothed_y, smoothed_a
)

MAX_CORRECTION_PX  = 30
MAX_CORRECTION_DEG = 1.0

correction_x = np.clip(correction_x, -MAX_CORRECTION_PX,  MAX_CORRECTION_PX)
correction_y = np.clip(correction_y, -MAX_CORRECTION_PX,  MAX_CORRECTION_PX)
correction_a = np.clip(correction_a, -MAX_CORRECTION_DEG, MAX_CORRECTION_DEG)

print("Phase 2 done")
print(f"Smooth radius    : {SMOOTH_RADIUS} frames")
print(f"Max x correction : {np.abs(correction_x).max():.2f}px")
print(f"Max y correction : {np.abs(correction_y).max():.2f}px")
print(f"Max a correction : {np.abs(correction_a).max():.4f}deg")
```

```python
def build_correction_matrix(dx, dy, angle_deg, frame_w, frame_h):
    center = (frame_w / 2, frame_h / 2)
    matrix = cv.getRotationMatrix2D(center, angle_deg, 1.0)

    matrix[0, 2] += dx
    matrix[1, 2] += dy
    
    return matrix


def crop_frame(frame, crop_ratio):
    h, w = frame.shape[:2]
    crop_x = int(w * crop_ratio)
    crop_y = int(h * crop_ratio)
    return frame[crop_y:h-crop_y, crop_x:w-crop_x]


def stabilise_video(video_path, output_path, correction_x, correction_y,
                    correction_a, crop_ratio=0.05, scale_preview=0.25):
    cap = cv.VideoCapture(video_path)
    fps    = cap.get(cv.CAP_PROP_FPS)
    width  = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
    total  = int(cap.get(cv.CAP_PROP_FRAME_COUNT))
    
    print(f"Input  : {width}x{height} @ {fps:.2f}fps  —  {total} frames")
    
    crop_x  = int(width  * crop_ratio)
    crop_y  = int(height * crop_ratio)
    out_w   = width  - 2 * crop_x
    out_h   = height - 2 * crop_y
    
    print(f"Output : {out_w}x{out_h}  (cropped {crop_ratio*100:.0f}% each edge)")
    
    fourcc = cv.VideoWriter_fourcc(*'mp4v')
    out    = cv.VideoWriter(output_path, fourcc, fps, (out_w, out_h))
    ret, first_frame = cap.read()
    out.write(crop_frame(first_frame, crop_ratio))
    
    preview_frames = []
    
    for i in range(1, total):
        ret, frame = cap.read()
        if not ret:
            break
        
        matrix = build_correction_matrix(
            correction_x[i-1],   
            correction_y[i-1],  
            correction_a[i-1],
            width, height
        )

        stabilised = cv.warpAffine(frame, matrix, (width, height),
                                   flags=cv.INTER_LINEAR,
                                   borderMode=cv.BORDER_REFLECT)

        stabilised_cropped = crop_frame(stabilised, crop_ratio)
        out.write(stabilised_cropped)
        
        # save a handful of frames for the preview plot
        if i in [50, 100, 150, 200]:
            original_cropped = crop_frame(frame, crop_ratio)
            preview_frames.append((i, original_cropped, stabilised_cropped))
        
        if i % 50 == 0:
            print(f"  {i}/{total} frames written")
    
    cap.release()
    out.release()
    print(f"\nDone! Saved to: {output_path}")
    
    return preview_frames


# ── run it ────────────────────────────────────────────────────────────────────
OUTPUT_PATH = "stabilised_output.mp4"

stabilise_video(
    VIDEO, OUTPUT_PATH,
    correction_x, correction_y, correction_a,
    crop_ratio=0.05
)
```
